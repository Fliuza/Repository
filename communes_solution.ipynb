```python
# 7. Compute the difference between the maximum and minimum elevations for each canton.
# Find the top 5 cantons that have the largest range of elevations?
```

```python
# define the grouping by canton and elevation, 
# and apply lambda funtion to get the difference between min and max. 
# Add reset_index to add range result as a new column.
df.groupby("Canton")["Elevation"].apply(lambda g: g.max() - g.min()).reset_index(name="Elevation_range")
```

```python
# get top 5 cantons with largest range of elevations
# add nlargest function and apply it to the new column 'elevation_range' to get top range values
df.groupby("Canton")["Elevation"].apply(lambda g: g.max() - g.min()).reset_index(name="Elevation_range").nlargest(5, "Elevation_range")
```

```python
# Define a data frame matrix whose rows correspond to communes and the columns to the cantons. 
# Fill in the matrix with 0/1 values where entry 
# (i,j) is a 1 if the commune in row  i is in the canton in column j and a 0 otherwise.
```

```python
# define relevant dataframe
df2=df[["Canton", "Commune"]]
```

```python
# using pandas get_dummies fucntion would be the simpliest way to obtain the desired result as the values are Categorical and no calculations are necessary to perform. 
# adding the groupby and applying the max will assign either 0 or 1 to the Commune that is related to appropriate canton. (get_dummies function discovered in pandas.pydata.org)
pd.get_dummies(df2.Canton).groupby(df2.Commune).apply(max)
```
